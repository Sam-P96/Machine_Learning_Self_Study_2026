```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("./Section 7 - Support Vector RegresSION (SVR)/Position_Salaries.csv") # Double check the folders if you get errors.

X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, -1].values

print(X)
print(y)
```

```python
y = y.reshape(len(y), 1)

print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train_linear, X_test_linear, y_train_linear, y_test_linear = train_test_split(X, y, test_size = 0.2, random_state = 1)
from sklearn.linear_model import LinearRegression
regressor_linear = LinearRegression()
regressor_linear.fit(X_train_linear, y_train_linear)
```

```python
plt.title("Linear Regression")
plt.scatter(X_train_linear, y_train_linear, color = 'red')
plt.plot(X_train_linear, regressor_linear.predict(X_train_linear), color='blue')
plt.show()

print("""Here you can see how linear regression yields a straight line, relatively
simple, easy to understand, but its kinda lacking the nuance cus its just a
straight line.""")
```

```python
# This is for multiple linear regression, I will write notes on this later.
```

```python
from sklearn.preprocessing import PolynomialFeatures
regressor_poly = PolynomialFeatures(degree = 2)
X_poly = regressor_poly.fit_transform(X)
poly_lin_reg = LinearRegression()
poly_lin_reg.fit(X_poly, y)
```

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, poly_lin_reg.predict(X_poly), color = 'blue')
plt.show()
print("""Polynomial regression at the 2nd degree yields a curve much closer to the scattered points""")
```

```python
regressor_poly = PolynomialFeatures(degree = 6)
X_poly = regressor_poly.fit_transform(X)
poly_lin_reg = LinearRegression()
poly_lin_reg.fit(X_poly, y)
plt.scatter(X, y, color = 'red')
plt.plot(X, poly_lin_reg.predict(X_poly), color = 'blue')
plt.show()
print("""By increasing thw degree, you can get it to match up closer""")
```
